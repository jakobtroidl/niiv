```python
import torch
import matplotlib.pyplot as plt
```

```python
gt_path = "/home/jakobtroidl/Desktop/neural-volumes/logs/hemibrain-volume-noisy-test-gradient-regularizer/results_iteration_41/15548_15559_15781/gt/gt_0.png"
gt_img = plt.imread(gt_path)
gt_img = torch.tensor(gt_img, dtype=torch.float32).view(1, 1, 128, 128)

reconstructed_path ="/home/jakobtroidl/Desktop/neural-volumes/logs/hemibrain-volume-noisy-test-gradient-regularizer/results_iteration_41/15548_15559_15781/result/result_0_psnr_21,788025427234373.png"
reconstructed_img = plt.imread(reconstructed_path)
reconstructed_img = torch.tensor(reconstructed_img, dtype=torch.float32).view(1, 1, 128, 128)

print(gt_img.shape)
print(reconstructed_img.shape)
```

```python

diff = torch.abs(gt_img - reconstructed_img)

# Display the original, denoised images, and the difference
plt.figure(figsize=(18, 6))

plt.subplot(131), plt.imshow(gt_img.squeeze().cpu().detach().numpy(), cmap='gray')
plt.title('Noisy GT'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(reconstructed_img.squeeze().cpu().detach().numpy(), cmap='gray')
plt.title('Reconstructed Image'), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(diff.squeeze().cpu().detach().numpy(), cmap='gray')
plt.title('Image Difference'), plt.xticks([]), plt.yticks([])

plt.show()
```
