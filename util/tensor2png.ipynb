```python
import numpy as np
import torch
import os
from PIL import Image
```

```python
input_path = '/home/jakobtroidl/Desktop/NVP/data/baselines/downsampled_128_128_128.npy'
output_path = '/home/jakobtroidl/Desktop/NVP/data/EM-small/'
```

```python
# load data
gt = np.load(input_path)

for i in range(gt.shape[2]):

    data = np.repeat(np.expand_dims(gt[:, :, i] * 255, axis=2), 3, axis=2)
    print(data.shape)
    im = Image.fromarray(data.astype('uint8'))
    name = output_path + str(i) + ".png"
    im.save(str(name))
```
